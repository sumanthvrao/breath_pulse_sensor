```python
import serial
import time
import datetime
from pymongo import MongoClient


# Configuration
serial_port = '/dev/ttyACM4'
period = 'night'

cnt = 0

# Connect to Serial Port for communication
ser = serial.Serial(serial_port, 115200, timeout=0)

# Connect to MongoDB
client = MongoClient('localhost:27017')
db = client.sensordata

fixed_interval = 0.3
while cnt < 100:
    try:
        d= ser.readline().strip()
        a=1
        t_tot_old=1
        t_tot_new=1
        #print(d)
        d=str(d)
        if(len(d)==0):
            continue
        if(d[:2]=='BP'):
            bpm = int(d[6:])
            doc_id = db.bp_data.insert_one({'datetime': datetime.datetime.now(),
                                             'bpm': bpm,'period':period}).inserted_id
            print(str(doc_id) + ': ' + str(datetime.datetime.now()) + ',' +str(bpm)+','+str(period))
            #print(datetime.datetime.now(),bpm)
        if(d=='IN'or d=='OUT'):
            doc_id = db.breath_data.insert_one({'datetime': datetime.datetime.now(),
                                             'type': d,'period':period}).inserted_id
            print(str(doc_id) + ': ' + str(datetime.datetime.now()) + ',' +d+','+str(period))
            cnt+=0.5     
    except serial.SerialTimeoutException:
        continue
    except ValueError:
        continue
        
        
```

```python
from pymongo import MongoClient
import numpy
# Connect to MongoDB
client = MongoClient('localhost:27017')
db = client.test1
L1=[['datetime','bpm']]
L2 = [['datetime','type']]


def read():
    try:
        xx = db.bp.find()
        print ('\n All data from  Database \n')
        for emp in xx:
            M=[]
            M.append(emp['datetime'])
            M.append(emp['bpm'])
            L1.append(M)
            
        yy = db.col_distance.find()    
        for emp in yy:
            N=[]
            N.append(emp['datetime'])
            N.append(emp['type'])
            L2.append(N)
         
        for i in L1:
            print(i)
        print("-----------------------------------------------------------")
        for i in L2:
            print(i)

    except Exception as e:
        print(e)
        
read()
```
